spec format is [[mz, intensity], [mz, intensity]]

assume precursor @ mz = 10 has been removed

```python
query = np.array([[1,0.1], [3,0.6], [4,0.3]])

target_1 = np.array([[1,0.6],[2,0.3],[3,0.1]])
target_2 = np.array([[5,0.1], [7,0.9]])
target_3 = np.array([[1,0.1], [3,0.9]])
```

Here are some ideas for match functions that take many inputs

```python
def closest_match(query, targets):
    """  
    Takes the closest match (abs dif in intensity)
    """

    composite_spec = list()
    for mz_q, int_q in query:

        #initialize match with intensity of 0
        match = [mz_q, 0]
        lowest_intensity_dif = 1

        for spec in targets:

            if mz_q in spec[:,0]:

                target_intensity = spec[np.where(spec[:,0] == mz_q)][0][1]

                #if the difference in intensities is the lowest we have seen so far, hang onto it
                if abs(int_q - target_intensity) < lowest_intensity_dif:

                    match[1] = target_intensity
                    lowest_intensity_dif = abs(int_q - target_intensity)

        composite_spec.append(match)

    return np.array(composite_spec)

def highest_intensity_match(query, targets):
    """  
    Takes the highest intensity for the same peak
    """

    composite_spec = list()
    for mz_q, int_q in query:

        #initialize match with intensity of 0
        match = [mz_q, 0]
        highest_intensity = 0

        for spec in targets:

            if mz_q in spec[:,0]:

                target_intensity = spec[np.where(spec[:,0] == mz_q)][0][1]

                #if this is the highest intensity we have seen, hang onto it
                if target_intensity > highest_intensity:

                    match[1] = target_intensity
                    highest_intensity = target_intensity

        composite_spec.append(match)

    return np.array(composite_spec)
```

```python
closest_match(query, [target_1, target_2, target_3])
```

```python
highest_intensity_match(query, [target_1, target_2, target_3])
```
